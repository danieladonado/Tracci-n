```python
#Importamos las librerías y cargamos el archivo de datos

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
df = pd.read_excel("AcerTaller_J8.xlsx")
```

```python
# Definimos las variables de esfuerzo y deformación
area = 0.00007126 #m²
long_ini= 50 #mm
esfuerzo=df['Fuerza (kN)']/area #kN/m² kilopascales
esfuerzo_pascales= esfuerzo/1e3 # Mega Pascales
deformacion = df['Cambio en la Long. (mm)']/long_ini #Adimensional
```

```python
# Graficamos la curva Tensión-Deformación
fig, ax = plt.subplots()

ax.plot(deformacion,esfuerzo_pascales)
ax.set_title("Curva Esfuerzo-Deformación", fontsize=13, color = "red")
ax.set_xlabel("Deformación unitaria")
ax.set_ylabel("Esfuerzo (MPa)")
plt.grid()
plt.xlim(0,1)
plt.ylim(0,600)
ax.scatter(0.1235, 447.3, color="red", zorder=5, label="Límite de elasticidad")
ax.text(0.15, 410, "Límite de elasticidad", rotation=0,
        fontfamily="Arial", color="red", va="center")


#Cálculo Módulo de Young
A=  (0.11,417.5)
B= (0.04,160)
m=(B[1]-A[1])/(B[0]-A[0])

#Límite de elasticidad
x = np.linspace(0, 1, 100)
b=10
y= m*(x-0.002)+ b
ax.plot(x, y, "--", color="red")

plt.show()



```
